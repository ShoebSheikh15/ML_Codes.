```python
df = df.rename(columns={
    'Mean of the integrated profile': 'MIP',
    'Standard deviation of the integrated profile': 'SDIP',
    'Excess kurtosis of the integrated profile': 'EKIP',
    'Skewness of the integrated profile': 'SKIP',
    'Mean of the DM-SNR curve': 'MDSNR',
    'Standard deviation of the DM-SNR curve': 'SDSNR',
    'Excess kurtosis of the DM-SNR curve': 'EKSNR',
    'Skewness of the DM-SNR curve': 'SKSNR',
    'target_class': 'Target'
})
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
svc=SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
accuracy_score(y_test,y_pred)
print('Model accuracy score with default hyperparameter: {0:0.4f}' .format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
 print('confusion_matrix')
cm=confusion_matrix(y_test, y_pred)
print(confusion_matrix(y_pred, y_test))
```

```python
print('accuracy_score')
print(accuracy_score(y_pred, y_test))
```

```python
print('classification_report')
print(classification_report(y_pred, y_test))
```

```python
svs=SVC(C=100.0)
svs.fit(X_train,y_train)
y_pred_svs = svs.predict(X_test)
print('Model accuracy score with rbf kernel and C=100.0 :{0:0.4f}'.format(accuracy_score(y_test, y_pred_svs)))
```

```python
linear_svc=SVC(kernel='linear',C=1.0)
linear_svc.fit(X_train,y_train)
y_pred_test=linear_svc.predict(X_test)
linear_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with linear kernel and c=1.0 : {0:.4f}'.format(linear_accuracy))
```

```python
linear_svc=SVC(kernel='linear',C=100.0)
linear_svc.fit(X_train,y_train)
y_pred_test=linear_svc.predict(X_test)
linear_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with linear kernel and c=100.0 : {0:.4f}'.format(linear_accuracy))
```

```python
linear_svc=SVC(kernel='linear',C=1000.0)
linear_svc.fit(X_train,y_train)
y_pred_test=linear_svc.predict(X_test)
linear_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with linear kernel and c=1000.0 : {0:.4f}'.format(linear_accuracy))
```

SVM with polynomial

```python
poly_svc=SVC(kernel='poly',C=1.0)
poly_svc.fit(X_train,y_train)
y_pred_test=poly_svc.predict(X_test)
poly_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with polynomial kernel and c=1.0 : {0:.4f}'.format(poly_accuracy))
```

```python
poly_svc=SVC(kernel='poly',C=100.0)
poly_svc.fit(X_train,y_train)
y_pred_test=poly_svc.predict(X_test)
poly_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with polynomial kernel and c=100.0 : {0:.4f}'.format(poly_accuracy))
```

svm with sigmoid

```python
sigmoid_svc=SVC(kernel='sigmoid',C=1.0)
sigmoid_svc.fit(X_train,y_train)
y_pred_test=sigmoid_svc.predict(X_test)
sigmoid_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with sigmoid kernel and c=1.0 : {0:.4f}'.format(sigmoid_accuracy))
```

```python
sigmoid_svc=SVC(kernel='sigmoid',C=100.0)
sigmoid_svc.fit(X_train,y_train)
y_pred_test=sigmoid_svc.predict(X_test)
sigmoid_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with sigmoid kernel and c=100.0 : {0:.4f}'.format(sigmoid_accuracy))
```

```python
sigmoid_svc=SVC(kernel='sigmoid',C=1000.0)
sigmoid_svc.fit(X_train,y_train)
y_pred_test=sigmoid_svc.predict(X_test)
sigmoid_accuracy = accuracy_score(y_test,y_pred_test)
print('Model accuracy score with sigmoid kernel and c=1000.0 : {0:.4f}'.format(sigmoid_accuracy))
```
